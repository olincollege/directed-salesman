```python
sizes = list(range(13))
sizes.remove(0)
sizes.remove(1)
sizes.remove(2)
num_runs = 30

times_held_karp, corr_held_karp = alg_test.run_multiple_on_range_with_correct(utils.random_graph, held_karp.held_karp, sizes, 1)
print("held karp done")
times_brute_force, corr_brute_force = alg_test.run_multiple_on_range_with_correct(utils.random_graph, brute_force.brute_force, sizes, 1)
print("brute force done")

times_nearest_neighbor, corr_nearest_neighbor = alg_test.run_multiple_on_range_with_correct(utils.random_graph, nearest_neighbor.tsp_nearest_neighbor, sizes, num_runs)
print("nearest neighbor done")

times_branch_and_bound, corr_branch_and_bound = alg_test.run_multiple_on_range_with_correct(utils.random_graph, branch_and_bound.branch_and_bound, sizes, num_runs)

```
